# 04 — Deep MLP Pipeline

Record-based moving window, interval-return labeling.

Core logic: `src/fqw/experiments/mlp.py`

```python
# !pip install -e ".."

import warnings
warnings.filterwarnings("ignore")

import pandas as pd
from fqw.config import MLPConfig
from fqw.experiments import run_mlp_for_ticker, sweep_mlp_thresholds

cfg = MLPConfig(data_dir="../data", alpha=0.55)
cfg.tickers
```

## Train + save probabilities

```python
TICKER = "GLDRUB_TOM"
probs = run_mlp_for_ticker(TICKER, config=cfg, use_technical_indicators=False)
probs.head() if probs is not None else None
```

## Threshold sweep

```python
if probs is not None:
    sweep_mlp_thresholds(probs, config=cfg)
```

## Batch run

```python
for ticker in cfg.tickers:
    run_mlp_for_ticker(ticker, config=cfg, use_technical_indicators=False)
```
