# 06 — Results Visualization

Scan `*_probs.parquet` files and plot thesis figures.

Core logic: `src/fqw/viz/results.py`

```python
# !pip install -e ".."

import warnings
warnings.filterwarnings("ignore")

from pathlib import Path
import matplotlib.pyplot as plt
import seaborn as sns

from fqw.viz import (
    scan_probability_results,
    plot_equity_curves,
    plot_metrics_comparison,
    plot_risk_return_scatter,
    plot_correlation_heatmap,
)

PLOTS_DIR = Path("../results/thesis_plots")
PLOTS_DIR.mkdir(parents=True, exist_ok=True)

PROJECT_DIRS = [
    ("../results/mlp", "MLP", "MLP"),
    ("../results", "CNN", "CNN"),
]

plt.rcParams.update({"figure.dpi": 150, "savefig.dpi": 300})
```

## Scan results

```python
metrics_df, equity_store, colors = scan_probability_results(PROJECT_DIRS)
metrics_df.head()
```

## Generate plots

```python
plot_equity_curves(equity_store, PLOTS_DIR, colors=colors)
plot_metrics_comparison(metrics_df, PLOTS_DIR, colors=colors)
plot_risk_return_scatter(metrics_df, PLOTS_DIR, colors=colors)
plot_correlation_heatmap(metrics_df, PLOTS_DIR)
print(f"Saved to {PLOTS_DIR}")
```
